```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
data = pd.read_csv("/content/Unemployment in India.csv")
data = pd.read_csv("/content/Unemployment_Rate_upto_11_2020.csv")
data
```

```python
data.describe()
```

```python
data.corr()
```

**DATA VISUALIZATION**

```python
plt.style.use('seaborn-darkgrid')
plt.figure(figsize=(15, 15))
sns.heatmap(data.corr())
plt.show()
```

**UMPLOYMENT RATE IN INDIA AS PER REGIONS**

```python
data.columns= ["States","Date","Frequency",
               "Estimated Unemployment Rate","Estimated Employed",
               "Estimated Labour Participation Rate","Region",
               "longitude","latitude"]

plt.figure(figsize=(15,15))
plt.title("Indian Unemployment")
sns.histplot(x="Estimated Unemployment Rate", hue="Region", data=data)
plt.show()
```

```python
sns.pairplot(data)
```

**PIE CHART**

```python
unp = data[["States", "Region", "Estimated Unemployment Rate"]]
figure = px.sunburst(unp, path=["Region", "States"], 
                     values="Estimated Unemployment Rate", 
                     width=700, height=700, color_continuous_scale="RdY1Gn", 
                     title=" Unemployment Rate in India")
figure.show()
```
